```python
import numpy as np

def marina(P):
  # global P

  no_answer = True

  length = len(P)
  for i in range(length):
    AB = [P[i][0] - P[(i-1)%length][0], P[i][1] - P[(i-1)%length][1]]
    BC = [P[(i+1)%length][0] - P[i][0], P[(i+1)%length][1] - P[i][1]]
    T = np.array(AB + BC).reshape(2, 2, order='F')

    signed_area = np.linalg.det(T)

    if signed_area < 0: 
      print(i+1)
      no_answer = False
  

  if no_answer: print(0)
```

```python
def test():
  for i in range(1, 4):
    print(f"[{i}]")
    inp_path = inp_pth(i)
    with open(inp_path, 'r') as inp:
      input = inp.readlines()
      
    P = input[1:]
    P = [list(map(int, p.split())) for p in P]
    

    print("[Out]")
    marina(P)

    out_path = out_pth(i)
    with open(out_path) as out:
      answer = out.read()
    
    print("[Answer]")
    print(answer)
```

```python
test()
```
